```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters, try out different values!!
num_points = 100    # Number of points in the simulation grid
num_steps = 1000    # Number of time steps
delta_x = 1.0       # Spatial step size
delta_t = 0.01      # Time step size
viscosity = 10     # Fluid viscosity 

# Initialize the grid with a sine wave as the initial velocity profile
x = np.linspace(0, num_points * delta_x, num_points)
u = np.sin(2 * np.pi * x / (num_points * delta_x))

# Function to update the grid using the 1D Burgers' equation
def update_grid(u, viscosity, delta_t, delta_x):
    u_new = np.copy(u)
    for i in range(1, num_points - 1):
        du_dx = (u[i + 1] - u[i - 1]) / (2 * delta_x)
        d2u_dx2 = (u[i + 1] - 2 * u[i] + u[i - 1]) / (delta_x ** 2)
        u_new[i] = u[i] + viscosity * delta_t * d2u_dx2 - u[i] * du_dx * delta_t
    return u_new

# Main simulation loop
for step in range(num_steps):
    u = update_grid(u, viscosity, delta_t, delta_x)

# Plot the final result
plt.plot(x, u)
plt.axhline(1, color = 'black', linestyle = '--')
plt.axhline(-1, color = 'black', linestyle = '--')

plt.xlabel('Position')
plt.ylabel('Velocity')
plt.title('1D Fluid Flow Simulation')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_points = 100    # Number of points in the simulation grid
num_steps = 1000    # Number of time steps
delta_x = 1.0       # Spatial step size
delta_t = 0.01      # Time step size
viscosity = 0.01     # Fluid viscosity

# Initialize the grid with two different initial velocity profiles
x = np.linspace(0, num_points * delta_x, num_points)

# Laminar flow profile (Sine wave)
initial_velocity_laminar = np.sin(2 * np.pi * x / (num_points * delta_x))

# Turbulent flow profile (Combination of sine waves)
initial_velocity_turbulent = np.sin(2 * np.pi * x / (num_points * delta_x)) + 0.5 * np.sin(4 * np.pi * x / (num_points * delta_x))

# Function to update the grid using the 1D Burgers' equation
def update_grid(u, viscosity, delta_t, delta_x):
    u_new = np.copy(u)
    for i in range(1, num_points - 1):
        du_dx = (u[i + 1] - u[i - 1]) / (2 * delta_x)
        d2u_dx2 = (u[i + 1] - 2 * u[i] + u[i - 1]) / (delta_x ** 2)
        u_new[i] = u[i] + viscosity * delta_t * d2u_dx2 - u[i] * du_dx * delta_t
    return u_new

# Main simulation loop for laminar flow
u_laminar = np.copy(initial_velocity_laminar)
for step in range(num_steps):
    u_laminar = update_grid(u_laminar, viscosity, delta_t, delta_x)

# Main simulation loop for turbulent flow
u_turbulent = np.copy(initial_velocity_turbulent)
for step in range(num_steps):
    u_turbulent = update_grid(u_turbulent, viscosity, delta_t, delta_x)
plt.axhline(1, color = 'black', linestyle = '--')
plt.axhline(-1, color = 'black', linestyle = '--')


    
# Plot the final results
plt.plot(x, u_laminar, label='Laminar Flow')
plt.plot(x, u_turbulent, label='Turbulent Flow')
plt.xlabel('Position')
plt.ylabel('Velocity')
plt.grid(True)
plt.legend()
plt.show()
```

This code is performing a Monte Carlo simulation of particle decay. The simulation models the decay of a collection of particles over a series of time steps, based on a given probability of decay for each particle in each time step. The code then plots the number of decayed particles over time to observe the decay process.

Breakdown:

1. Import necessary libraries: `numpy` for numerical computations and `matplotlib.pyplot` for plotting the results.

2. Define simulation parameters:
   - `num_particles`: The total number of particles in the simulation.
   - `decay_prob`: The probability of particle decay in a single time step. This probability is used to determine whether each particle decays or not in each time step.
   - `num_steps`: The number of time steps for the simulation.

3. Initialize the `particles` array: An array of length `num_particles` is created to represent the status of each particle. Each element of the array is initialized to 0, indicating that all particles are undecayed initially.

4. Define the `simulate_decay` function: This function simulates the decay process for a given set of particles and decay probability. It goes through each particle and checks if it is undecayed (represented by 0). If the particle is undecayed, it uses `np.random.rand()` to generate a random number between 0 and 1. If this random number is less than the `decay_prob`, the particle is marked as decayed (1), and the function keeps track of the number of decayed particles. The function returns the count of decayed particles.

5. Main simulation loop: The code runs a loop for `num_steps` time steps. In each step, the `simulate_decay` function is called to simulate particle decay, and the count of decayed particles in that time step is appended to the `decayed_counts` list.

6. Plot the decayed particle counts over time: The final step is to plot the number of decayed particles over time using `matplotlib.pyplot`. The x-axis represents the time steps, and the y-axis represents the number of decayed particles at each time step. The plot includes labels for the axes, a title, and grid lines for better visualization.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_particles = 1000    # Number of particles in the simulation
decay_prob = 0.05       # Probability of particle decay in a time step
num_steps = 100         # Number of time steps

# Arrays to store the particle status (0 for undecayed, 1 for decayed)
particles = np.zeros(num_particles, dtype=int)

# Function to simulate particle decay using Monte Carlo method
def simulate_decay(particles, decay_prob):
    decayed_particles = 0
    for i in range(len(particles)):
        if particles[i] == 0:
            if np.random.rand() < decay_prob:
                particles[i] = 1
                decayed_particles += 1
    return decayed_particles

# Main simulation loop
decayed_counts = []
for step in range(num_steps):
    decayed_particles = simulate_decay(particles, decay_prob)
    decayed_counts.append(decayed_particles)

# Plot the decayed particle counts over time
time_steps = np.arange(num_steps)
plt.plot(time_steps, decayed_counts)
plt.xlabel('Time Steps')
plt.ylabel('Decayed Particles')
plt.title('Monte Carlo Simulation of Particle Decay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_particles = 1000    # Number of particles in the simulation
decay_prob = 0.05       # Probability of particle decay in a time step
num_steps = 100         # Number of time steps
num_simulations = 100   # Number of Monte Carlo simulations

# Function to perform the Monte Carlo simulation
def simulate_decay(num_particles, decay_prob, num_steps, num_simulations):
    decay_rates = np.zeros((num_simulations, num_steps))
    for sim in range(num_simulations):
        # Generate random numbers for all particles and time steps at once
        random_numbers = np.random.rand(num_particles, num_steps)
        
        # Initialize arrays to store particle status (0 for undecayed, 1 for decayed)
        particles = np.zeros((num_particles, num_steps), dtype=int)
        
        # Simulate particle decay
        particles[random_numbers < decay_prob] = 1
        
        # Calculate the decay rate over time for this simulation
        decay_rates[sim] = np.mean(particles, axis=0)
    
    # Calculate the mean decay rate and confidence intervals over all simulations
    mean_decay_rate = np.mean(decay_rates, axis=0)
    lower_bound = np.percentile(decay_rates, 2.5, axis=0)
    upper_bound = np.percentile(decay_rates, 97.5, axis=0)
    
    return mean_decay_rate, lower_bound, upper_bound

# Perform the Monte Carlo simulation
mean_decay_rate, lower_bound, upper_bound = simulate_decay(num_particles, decay_prob, num_steps, num_simulations)

# Plot the mean decay rate and confidence intervals over time
time_steps = np.arange(num_steps)
plt.plot(time_steps, mean_decay_rate, label='Mean Decay Rate', color='b')
plt.fill_between(time_steps, lower_bound, upper_bound, color='b', alpha=0.3, label='95% Confidence Interval')
plt.axhline(y=decay_prob, color='r', linestyle='--', label='Actual Decay Probability')
plt.xlabel('Time Steps')
plt.ylabel('Decay Rate')
plt.title('Monte Carlo Simulation of Particle Decay')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Acceleration due to gravity (m/s^2)

# Function to simulate projectile motion
def simulate_projectile(v0, theta, dt, num_steps):
    # Initialize arrays to store time, position, and velocity
    t = np.linspace(0, dt * num_steps, num_steps + 1)
    x = np.zeros(num_steps + 1)
    y = np.zeros(num_steps + 1)
    vx = v0 * np.cos(np.radians(theta))
    vy = v0 * np.sin(np.radians(theta))

    # Perform numerical integration
    for i in range(num_steps):
        x[i + 1] = x[i] + vx * dt
        y[i + 1] = y[i] + vy * dt
        vy -= g * dt

    # Store data in a Pandas DataFrame
    data = pd.DataFrame({"Time (s)": t, "X position (m)": x, "Y position (m)": y})
    return data

# Input parameters
initial_velocity = 20.0  # m/s
launch_angle = 45.0  # degrees
time_step = 0.01  # s
num_steps = 1000

# Run the simulation
projectile_data = simulate_projectile(initial_velocity, launch_angle, time_step, num_steps)

# Plot the trajectory
plt.figure()
plt.plot(projectile_data["X position (m)"], projectile_data["Y position (m)"])
plt.xlabel("X position (m)")
plt.ylabel("Y position (m)")
plt.title("Projectile Motion")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Acceleration due to gravity (m/s^2)
rho = 1.225  # Air density at sea level (kg/m^3)
Cd = 0.4  # Drag coefficient (spherical projectile)
A = 0.01  # Cross-sectional area of the projectile facing motion (m^2)
m = 0.1  # Mass of the projectile (kg)

# Function to simulate projectile motion with air resistance
def simulate_projectile_air_resistance(v0, theta, dt, num_steps):
    # Initialize arrays to store time, position, and velocity
    t = np.linspace(0, dt * num_steps, num_steps + 1)
    x = np.zeros(num_steps + 1)
    y = np.zeros(num_steps + 1)
    vx = v0 * np.cos(np.radians(theta))
    vy = v0 * np.sin(np.radians(theta))

    # Perform numerical integration
    for i in range(num_steps):
        v = np.sqrt(vx ** 2 + vy ** 2)
        ax = -(Cd * rho * A * v * vx) / (2 * m)
        ay = -g - (Cd * rho * A * v * vy) / (2 * m)

        x[i + 1] = x[i] + vx * dt
        y[i + 1] = y[i] + vy * dt

        vx += ax * dt
        vy += ay * dt

    # Store data in a Pandas DataFrame
    data = pd.DataFrame({"Time (s)": t, "X position (m)": x, "Y position (m)": y})
    return data

# Input parameters
initial_velocity = 20.0  # m/s
launch_angle = 45.0  # degrees
time_step = 0.01  # s
num_steps = 1000

# Run the simulation with air resistance
projectile_data_air_resistance = simulate_projectile_air_resistance(
    initial_velocity, launch_angle, time_step, num_steps
)

# Plot the trajectory with air resistance
plt.figure()
plt.plot(
    projectile_data_air_resistance["X position (m)"],
    projectile_data_air_resistance["Y position (m)"],
    label="With Air Resistance",
)

# Run the simulation without air resistance for comparison
projectile_data_no_air_resistance = simulate_projectile(
    initial_velocity, launch_angle, time_step, num_steps
)

# Plot the trajectory without air resistance
plt.plot(
    projectile_data_no_air_resistance["X position (m)"],
    projectile_data_no_air_resistance["Y position (m)"],
    label="No Air Resistance",
)

plt.xlabel("X position (m)")
plt.ylabel("Y position (m)")
plt.title("Projectile Motion with Air Resistance")
plt.legend()
plt.grid()
plt.show()
```

```python

def add(x, y):
    """
    Function to add two numbers.

    Parameters:
        x (float): The first number.
        y (float): The second number.

    Returns:
        float: The sum of the two numbers.
    """
    return x + y

def subtract(x, y):
    """
    Function to subtract two numbers.

    Parameters:
        x (float): The first number.
        y (float): The second number.

    Returns:
        float: The difference between the two numbers.
    """
    return x - y

def multiply(x, y):
    """
    Function to multiply two numbers.

    Parameters:
        x (float): The first number.
        y (float): The second number.

    Returns:
        float: The product of the two numbers.
    """
    return x * y

def divide(x, y):
    """
    Function to divide two numbers.

    Parameters:
        x (float): The first number (dividend).
        y (float): The second number (divisor).

    Returns:
        float: The result of division (quotient).
    """
    return x / y

def calculator():
    """
    Simple text-based calculator.

    The user can select the desired operation (addition, subtraction,
    multiplication, or division) by entering the corresponding number.
    The program will then prompt for two numbers and perform the chosen operation
    using the appropriate function. The result will be displayed to the user.
    """
    print("Simple Calculator")
    print("Operations: ")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    operation = input("Enter the operation number (1/2/3/4): ")

    if operation in ['1', '2', '3', '4']:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        if operation == '1':
            result = add(num1, num2)
        elif operation == '2':
            result = subtract(num1, num2)
        elif operation == '3':
            result = multiply(num1, num2)
        else:
            result = divide(num1, num2)

        print("Result:", result)
    else:
        print("Invalid operation number. Please try again.")

if __name__ == "__main__":
    calculator()
```

## Calculator with addional features

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

def calculator():
    print("Advanced Calculator")
    print("Operations: ")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    while True: ## what is this doing? is it infinite?
        operation = input("Enter the operation number (1,2,3,4) or '5' to exit: ")

        if operation == '5':
            print("Goodbye!")
            break

        if operation in ['1', '2', '3', '4']:
            try:
                num1 = float(input("Enter the first number: "))
                num2 = float(input("Enter the second number: "))

                if operation == '1':
                    result = add(num1, num2)
                elif operation == '2':
                    result = subtract(num1, num2)
                elif operation == '3':
                    result = multiply(num1, num2)
                else:
                    result = divide(num1, num2)

                print("Result:", result)
            except ValueError:
                print("Invalid input. Please enter valid numbers.")
        else:
            print("Invalid operation number. Please try again.")

if __name__ == "__main__":
    calculator()
```

```python
import math
import ast

def calculator():
    print("Advanced Calculator")
    print("Supported Operations: +, -, *, /, ** (power), sin, cos, tan")
    print("Enter 'exit' to quit the calculator.")

    while True:
        expression = input("Enter an expression: ")

        if expression.lower() == 'exit':
            print("Goodbye!")
            break

        try:
            # Using ast.literal_eval() to safely evaluate expressions
            result = ast.literal_eval(expression)
            print("Result:", result)
        except (ValueError, ZeroDivisionError):
            print("Invalid expression or division by zero. Please try again.")
        except Exception as e:
            print("An error occurred:", e)

if __name__ == "__main__":
    calculator()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plume_simulation(mass_eruption_rate):
    return (mass_eruption_rate/2500) ** 0.25

# Mass eruption rates in kg/s
mass_eruption_rates = np.logspace(3, 10, 100)  # Logarithmic scale from 10^3 to 10^10
plume_heights = plume_simulation(mass_eruption_rates)

# Provided data: Volcano, Plume Height (km), Mass Eruption Rate (kg/s)
data = [
    ["St. Helens, 2005", 9, 4.00e5],
    ["St. Helens, 7/1980", 10.3, 1.40e6],
    ["St. Helens, 5/1980", 10.2, 2.00e7]
]

# Extract volcano names, heights, and eruption rates from the data
volcanoes, heights, eruption_rates = zip(*data)

# Convert heights from km to meters
heights_meters = [height * 1000 for height in heights]

# Create a custom color map
cmap = plt.get_cmap('viridis')

# Create the scatter plot with custom color map
plt.scatter(mass_eruption_rates, plume_heights, c=plume_heights, cmap=cmap, edgecolors='k', label='Data Points')
plt.colorbar(label='Plume Height (meters)')

# Annotate the data points with their values
for mass, height in zip(mass_eruption_rates[::10], plume_heights[::10]):
    plt.annotate(f"{mass:.1e} kg/s", xy=(mass, height), xytext=(mass * 1.5, height * 0.9), arrowprops=dict(arrowstyle="->", lw=0.5), fontsize=7)

# Add vertical lines to mark specific mass eruption rates
specific_mass_rates = [1e3, 1e5, 1e9]
for rate in specific_mass_rates:
    plt.axvline(rate, linestyle='--', color='gray', lw=2)

plt.xscale('log')
plt.xlabel("Mass Eruption Rate (kg/s)")
plt.ylabel("Plume Height (meters)")
plt.title("Volcanic Plume Height vs. Mass Eruption Rate")

# Customize the background grid style
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# Calculate and plot the provided data on top of the current plot
provided_eruption_rates = [rate for _, _, rate in data]
provided_plume_heights = [plume_simulation(rate) for _, _, rate in data]
plt.scatter(provided_eruption_rates, provided_plume_heights, c='red', marker='o', s=150, label='Provided Data')

plt.legend()
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
target_sum = 7         # Target sum for the player to win
num_simulations = 1000  # Number of Monte Carlo simulations
num_rounds = 1000      # Number of rounds in each simulation

# Function to simulate one round of the dice game
def play_dice_game():
    dice1 = np.random.randint(1, 7)
    dice2 = np.random.randint(1, 7)
    sum_of_dice = dice1 + dice2
    return sum_of_dice == target_sum

# Monte Carlo simulation
results = []
for _ in range(num_simulations):
    rounds_won = 0
    for _ in range(num_rounds):
        if play_dice_game():
            rounds_won += 1
    win_probability = rounds_won / num_rounds
    results.append(win_probability)

# Calculate the mean probability of winning
mean_win_probability = np.mean(results)

# Plot the probability of winning against the number of rounds played
rounds_played = np.arange(1, num_simulations + 1)
plt.plot(rounds_played, results)
plt.axhline(y=mean_win_probability, color='r', linestyle='--', label='Mean Win Probability')
plt.xlabel('Number of Rounds Played')
plt.ylabel('Probability of Winning')
plt.title('Monte Carlo Simulation of Casino Dice Game')
plt.legend()
plt.grid(True)
plt.show()

print(f"Estimated Probability of Winning: {mean_win_probability:.4f}")
```

In this simulation, we consider a simplified version of roulette with a single-zero wheel. The target_number variable represents the number the player bets on. The simulation runs num_simulations simulations, with each simulation consisting of num_spins spins of the roulette wheel. The player bets the same bet_amount on each spin.

The bet_outcome function determines whether the player wins the bet for a single spin based on the bet type, which can be a specific number, red/black, odd/even, etc. The payouts for the different bet types are not considered in this simplified version.


This simulation should assume a simplified version of roulette. If we wanted to design a more realistic one, we would have to  can consider real-world casino rules and payout schemes. For this assigment lets stick to simplicty and plot the results! Does the house always win??


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_simulations = 1000  # Number of Monte Carlo simulations
num_spins = 100        # Number of spins in each simulation
bet_amount = 10        # Amount of money bet on each spin
target_number = 17     # Number to bet on (e.g., red/black, odd/even)

# simulate one spin of the roulette wheel
def spin_roulette_wheel():
    return np.random.randint(0, 37)  # 0-36 represent the numbers on the wheel

# determine the outcome of a bet on a single spin
def bet_outcome(spin_result, bet_type):
    if bet_type == 'number':
        return spin_result == target_number
    elif bet_type == 'red':
        return spin_result % 2 == 1 and spin_result != 0
    elif bet_type == 'black':
        return spin_result % 2 == 0 and spin_result != 0
    elif bet_type == 'odd':
        return spin_result % 2 == 1
    elif bet_type == 'even':
        return spin_result % 2 == 0 and spin_result != 0

    
# Monte Carlo sim
results = []
for _ in range(num_simulations):
    total_winnings = 0
    for _ in range(num_spins):
        spin_result = spin_roulette_wheel()
        if bet_outcome(spin_result, 'number'):
            total_winnings += 36 * bet_amount  # 36:1 payout for betting on a single number
    results.append(total_winnings)

# Calculate the mean winnings
mean_winnings = np.mean(results)

# Plot the histogram of winnings
plt.hist(results, bins=30, density=True, alpha=0.7)
plt.axvline(x=mean_winnings, color='r', linestyle='--', label='Mean Winnings')
plt.xlabel('Total Winnings')
plt.ylabel('Probability Density')
plt.title('Monte Carlo Simulation of Roulette Game')
plt.legend()
plt.grid(True)
plt.show()

print(f"Estimated Mean Winnings: ${mean_winnings:.2f}")
```
