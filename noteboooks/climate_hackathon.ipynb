```python
state = 'Name'
burden = 'Energy Burden (% income)'
cost = 'Avg. Annual Energy Cost ($)'
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

sns.scatterplot(data = df,
                x = 'Name',
                y = cost,
                hue = 'Geography ID',
                palette = 'prism'
               )  # S
```

```python
g = sns.PairGrid(df[[state, burden, cost, "Household Income"]], hue="Name")
g.map_upper(sns.regplot)
g.map_lower(sns.residplot)
g.map_diag(plt.hist)
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)
g.add_legend()
g.set(alpha=0.5)
```
